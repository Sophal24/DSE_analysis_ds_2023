## Combine QuestionStats

```python
excel_data_paths = glob.glob('data_source/**/*.xlsx', recursive=True)


combined_df = pd.DataFrame()
tmp_combined_df = pd.DataFrame()
is_first_loop = True

for path in excel_data_paths:
    if('questionstats' in path):
        print(path)

        df_q_state = pd.read_excel(path, sheet_name=1)
        df_q_state.columns = df_q_state.columns.str.lower()

        facility_index_rate = []
        for facility_index in df_q_state['facility index']:
            facility_index = facility_index[:-1]
            # print(facility_index)
            facility_index_rate.append(float(facility_index))

        df_q_state['facility index'] = facility_index_rate

        df_q_state.loc[df_q_state['facility index'] <= float(60), 'label_facility_index'] = 'L3'
        df_q_state.loc[(df_q_state['facility index'] > float(60)) & (df_q_state['facility index'] <= float(80)), 'label_facility_index'] = 'L2'
        df_q_state.loc[(df_q_state['facility index'] > float(80)) & (df_q_state['facility index'] <= float(100)), 'label_facility_index'] = 'L1'

        # df = df_q_state
        df_q_state['name'] = ''

        names = []
        for name in df_q_state['question name']:
            if('Random' in name):
                names.append('')
            else:
                names.append(name)
        
        df_q_state['name'] = names

        df_q_state = df_q_state[['name','attempts','facility index', 'label_facility_index']]
        df_q_state = df_q_state[df_q_state['name'] != '']
        # print(df_q_state)

        if(is_first_loop):
            is_first_loop = False
            tmp_combined_df = df_q_state
            combined_df = tmp_combined_df
        else:
            next_df = df_q_state
            combined_df = pd.concat([combined_df, next_df], ignore_index=True)

# combined_df.to_excel('combined_question_stats.xlsx')
combined_df['name'].nunique()
```

```python
combined_df
```

```python
uniques_name = combined_df['name'].unique()

final_df_q_stats = pd.DataFrame()
tmp_combined_df = pd.DataFrame()
is_first_loop = True

for name in uniques_name:
    each_name_df = combined_df[combined_df['name'] == name]
    # print(each_name_df) #TODO

    value_count = each_name_df['label_facility_index'].value_counts(normalize=True)

    attempt = each_name_df['attempts'].sum()
    # print(attempt)

    correct_rate = each_name_df['facility index'].mean()
    # print(correct_rate)

    if(correct_rate <= float(60)):
        label = 'L3'
    elif(correct_rate > float(60) and correct_rate <= (80)):
        label = 'L2'
    elif(correct_rate > float(80) and correct_rate <= float(100)):
        label = 'L1'

    data = [[name,
             attempt,
             correct_rate,
             label]]

    if(is_first_loop):
        is_first_loop = False
        tmp_combined_df = pd.DataFrame(data, columns=['name',
                                                      'attempt',
                                                      'correct_rate',
                                                      'level'])
        final_df_q_stats = tmp_combined_df
    else:
        next_df = pd.DataFrame(data, columns=['name', 
                                              'attempt',
                                              'correct_rate',
                                              'level'])
        final_df_q_stats = pd.concat([final_df_q_stats, next_df], ignore_index=True)
        
    # break

# final_df_q_stats.to_excel('final_question_stats_with_label.xlsx')

final_df_q_stats
```

## Combine Question Pools

```python
xml = sorted(glob.glob('./data_source_xml/*.xml'))

combined_df_q_pool = pd.DataFrame()
tmp_combined_df_q_pool = pd.DataFrame()
is_first_loop = True

for path in xml:
    with open(path) as fd:
        doc = xmltodict.parse(fd.read())

    df_xml1 = pd.DataFrame(doc['quiz'])
    # df_xml1.head(3)

    columns_name = df_xml1['question'][0].keys()

    data = []
    subject_xml1_df = pd.DataFrame(data, columns=columns_name)

    for i in range(len(df_xml1)):
        subject_xml1_df.loc[i] = df_xml1['question'][i]

    names = []
    for i in range(len(subject_xml1_df)):
        names.append(subject_xml1_df['name'][i]['text'])

    subject_xml1_df['name'] = names
    # subject_xml1_df.head(3)

    level = []
    q_type = []
    for i in range(len(subject_xml1_df)):
        
        if(pd.isna(subject_xml1_df['tags'][i])):
            level.append('pending')
            q_type.append('pending')
        else:
            level_object = str(subject_xml1_df['tags'][i]['tag'])

            if('l1' in level_object or 'L1' in level_object):
                level.append('L1')
            elif('l2' in level_object or 'L2' in level_object):
                level.append('L2')
            elif('l3' in level_object or 'L3' in level_object):
                level.append('L3')
            elif('pending' in level_object or 'Pending' in level_object):
                level.append('pending')
            else:
                level.append('reviewing')
            
            if('exam' in level_object or 'Exam' in level_object):
                q_type.append('exam')
            elif('quiz' in level_object or 'Quiz' in level_object):
                q_type.append('quiz')
            elif('pretest' in level_object or 'Pretest' in level_object):
                q_type.append('pretest')
            elif('pending' in level_object or 'Pending' in level_object):
                q_type.append('pending')
            elif('review' in level_object or 'Review' in level_object):
                q_type.append('review')
            else:
                q_type.append(None)

    subject_xml1_df['level'] = level
    subject_xml1_df['q_type'] = q_type
    subject_xml1_df['subject_th'] = path.split('/')[-1].split('.')[0]
    
    if(is_first_loop):
        is_first_loop = False
        tmp_combined_df_q_pool = subject_xml1_df[['name', 'level', 'q_type','subject_th']]
        combined_df_q_pool = tmp_combined_df_q_pool
    else:
        next_df = subject_xml1_df[['name', 'level', 'q_type','subject_th']]
        combined_df_q_pool = pd.concat([combined_df_q_pool, next_df], ignore_index=True)
        
combined_df_q_pool.to_excel('./data_source_xml/combined_question_pool_total.xlsx')

question_with_label_only_df = combined_df_q_pool[combined_df_q_pool['level'].isin(['L1', 'L2', 'L3'])]
question_with_label_only_df.to_excel('./data_source_xml/question_pool_level_only.xlsx')

print("Combined dataset of question pools!!!")
print("Saved files to folders!")

print("Number of Total Question:", len(combined_df_q_pool))
print("Number of question with Label:", len(combined_df_q_pool[combined_df_q_pool['level'].isin(['L1', 'L2', 'L3'])]))
```
