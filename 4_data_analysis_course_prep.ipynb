```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

## CADT 

```python
df_cadt = pd.read_csv('./course_prep_dataset/progress_dse_cadt.csv')
df_cadt.head(3)
```

```python
columns = df_cadt.columns
for col in columns:
    if('Unnamed' in col):
        # print(col)
        df_cadt = df_cadt.drop(columns=[col], axis=1)
        
```

```python
new_name = 0
columns = df_cadt.columns
for col in columns:
    df_cadt.rename(columns={col: new_name}, inplace=True)
    new_name += 1
```

```python
# Replace True/False with Yes/No
df_cadt = df_cadt.replace({'Not completed': 0, 'Completed': 1})
```

```python
df_cadt = df_cadt.dropna(axis=1, how='all')
```

```python
df_cadt = df_cadt.rename({0:'email'}, axis=1)
df_cadt = df_cadt.drop(columns=['email'])
df_cadt
```

```python
column_sum = df_cadt.sum()
column_sum = column_sum.tolist()
height = column_sum
print('Height:', height )

label = []
number = 1
for h in height:
    label.append(number)
    number +=1 

print('Label', label)
```

```python
fig = plt.figure(figsize = (11, 5))
 
# creating the bar plot
plt.bar(label, height)
 

plt.ylim(top=len(df_cadt))
plt.xlabel("Slide and Videos lessons")
plt.ylabel("No. of examinee study pre-course")
plt.title("Number of examinees who completed slides and video lesson")
plt.show()
```

```python
fig = plt.figure(figsize = (11, 5))
 
# creating the bar plot
plt.plot(label, height)
 

plt.ylim(top=len(df_cadt))
plt.xlabel("Slide and Video Lessons")
plt.ylabel("No. of examinee study pre-course")
plt.title("Number of examinees who completed slides and video lesson")
plt.show()
```

```python
# sns.set(rc={'figure.figsize': (10, 5)})
# sns.lineplot(x=label, y=height)
```

```python
df_cadt['num_completed_lesson'] = df_cadt.sum(axis=1)
df_cadt['completed_rate'] = df_cadt['num_completed_lesson']/len(df_cadt.columns)
df_cadt.head(5)
```

```python
average = df_cadt['num_completed_lesson'].mean()
print("Average Completed Lessons:", average)

max = df_cadt['num_completed_lesson'].max()
print("Max:", max)

median = df_cadt['num_completed_lesson'].median()
print("Median:", median)

min = df_cadt['num_completed_lesson'].min()
print("Min:", min)

std = df_cadt['num_completed_lesson'].std()
print("Standard Deviation:", std)
```

```python
average = df_cadt['completed_rate'].mean()
average
```

```python
plt.figure(figsize=(12, 8))

completed_rate = average
not_completed_rate = 1-average

# Data for the pie chart
labels = 'Completed Rate', 'Not Completed Rate'
sizes = [completed_rate, not_completed_rate]

# Create the pie chart
plt.pie(sizes, labels=labels, autopct='%1.0f%%')

# Set the legend
plt.legend(title="Rate", loc='upper left')

# Show the plot
plt.show()
```
