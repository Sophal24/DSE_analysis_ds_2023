```python
act_completion_files = glob.glob('./course_prep_dataset/all_activity_completion_excel/*')
print("There are", len(act_completion_files), "files.")

print("Reading files and combine ...\n")
combine_act_completion_df = pd.DataFrame()
for file in act_completion_files:
    # print(file)

    df = pd.read_excel(file)
    df.rename(columns={df.columns[0]: 'name'}, inplace=True)

    columns = df.columns
    for col in columns:
        if('Unnamed' in col):
            df = df.drop(columns=[col], axis=1)

    new_name = 0
    columns = df.columns
    for col in columns:
        df.rename(columns={col: new_name}, inplace=True)
        new_name += 1

    df.rename(columns={df.columns[0]: 'name'}, inplace=True)
    df.rename(columns={df.columns[1]: 'email'}, inplace=True)

    df = df.replace({'Not completed': 0, 'Completed': 1})

    df['total_lesson'] = len(df.iloc[:, 2:].columns)

    df['completed_lesson'] = df.iloc[:, 2:-1].sum(axis=1)
    df['completed_rate'] = df['completed_lesson']/df['total_lesson']
    df['completed_rate_percentage'] = (df['completed_rate'] * 100).round(2)

    df = df[['name', 'email', 'total_lesson', 'completed_lesson', 'completed_rate', 'completed_rate_percentage']]
    # print(df)

    # combine dataset - all_pretest
    combine_act_completion_df = pd.concat([combine_act_completion_df, df], ignore_index=True)

combine_act_completion_df
```

```python
df_mapped = pd.read_excel('./course_prep_dataset/improvement_list_mapped.xlsx')
df_mapped
```

```python
mapped_name = np.array(df_mapped['name'])
df = df[df['name'].isin(mapped_name)]
df['pre_test_score'] = None
df['certified_score'] = None
df
```

```python
for name in np.array(df.name):
    pre_test_score = df_mapped[df_mapped['name'] == name]['pre_test_score'].values[0]
    certified_score = df_mapped[df_mapped['name'] == name]['Grade/100.00'].values[0]

    # print(pre_test_score)
    # print(certified_score)
    # print()

    df.loc[df['name'] == name, 'pre_test_score'] = pre_test_score
    df.loc[df['name'] == name, 'certified_score'] = certified_score
```

```python
df  = df.sort_values(by='name')
df.loc[df['certified_score'] >= float(60), 'result'] = 'Passed'
df.loc[df['certified_score'] < float(60), 'result'] = 'Failed'
df
```
